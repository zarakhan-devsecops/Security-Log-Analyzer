```python
logs = [
    "2026-06-26 10:01 Failed login from 192.168.1.10",
    "2026-06-26 10:02 Failed login from 192.168.1.10",
    "2026-06-26 10:03 Failed login from 192.168.1.10",
    "2026-06-26 10:05 Successful login from 192.168.1.20",
    "2026-06-26 10:07 Failed login from 192.168.1.30"
]

print("Security Logs Loaded Successfully")
```

```python
failed_logins = 0

for log in logs:
    if "Failed login" in log:
        failed_logins += 1

print("Total Failed Login Attempts:", failed_logins)
```

```python
if failed_logins >= 3:
    print("ALERT: Possible Brute Force Attack Detected!")
else:
    print("No Suspicious Activity Found")
```

```python
ip_counts = {}

for log in logs:
    if "Failed login" in log:
        ip = log.split()[-1]

        if ip in ip_counts:
            ip_counts[ip] += 1
        else:
            ip_counts[ip] = 1

print("Suspicious IP Addresses")

for ip, count in ip_counts.items():
    if count >= 3:
        print(ip, "-", count, "failed attempts")
```
